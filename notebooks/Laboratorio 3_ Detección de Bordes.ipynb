Encontrar los bordes de una imagen es una de las operaciones de procesamiento de imágenes más fundamentales. Los bordes pueden proporcionar pistas sobre los elementos y el significado de una imagen. Físicamente, un borde es el límite exterior de un objeto, área o superficie. Pero, ¿Qué es un borde dentro de una imagen 2D? Los bordes son cambios locales significativos en la imagen en los que el brillo cambia drásticamente. Estos cambios representan el límite entre dos o más regiones o superficies físicas materialmente distintas.

Por otro lado, la detección de bordes es una tarea difícil pues estos no siempre son claros debido a los rayos de luz y las sombras. Para ello, se han investigado muchos métodos desde los años 80. En este tutorial, se implementará un conjunto preliminar de métodos de detección de bordes que se basan únicamente en filtros.
